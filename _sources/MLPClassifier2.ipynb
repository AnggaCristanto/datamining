# MLPClasssifier Dataset Alkhohol2

### Masukkan librari

```python
import numpy as np # linear algebra
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

### Load File Dataset Iris Dengan CSV


```python
df = pd.read_csv('https://raw.githubusercontent.com/AnggaCristanto/dataset/main/alkhohol-2.csv', sep = ';')
df
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
y = df['Target']
x = df.drop(['Target'], axis=1)
```

### Memisahkan Data menjadi Data Latih dan Data Uji

### Melatih Model MLPClassifier

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=27)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(300,300,300), max_iter=1000, alpha=0.0001,
                     solver='sgd', verbose=10,  random_state=21,tol=0.001)
```

### Menguji Model Dan Untuk Menghitung Akurasi

```python
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(y_test, y_pred)
```
